```python
from pathlib import Path

# Find repo root
REPO_ROOT = Path.cwd().parent
print(f"Repo root: {REPO_ROOT}")

REPORT_ROOT = REPO_ROOT / "report"

FIGSIZE = (16,10)
DPI = 300
GENERATE_PLOTS = False
```

```python
DATA_ROOT = REPO_ROOT / "data"

import pandas as pd
# load xlsx file

median_income = pd.read_excel(DATA_ROOT / "raw" / "Medianeinkommen_Berlin-2023.xlsx")

import geopandas as gpd
# load lor shapes from shp

lor = gpd.read_file(DATA_ROOT / "raw" / "lor_2021-01-01_k3_shapefiles_nur_id" / "LOR_2023-01-01_PLR_EPSG_25833_nur_ID.shp")
```

```python
lor
```

```python
# skip first 5 rows and use 5th row as header
median_income = median_income.iloc[5:]
median_income.columns = median_income.iloc[0]
median_income = median_income.iloc[1:]

# remove all rows after row 547
median_income = median_income.iloc[:(547-5)]

# Reset index
median_income = median_income.reset_index(drop=True)
```

```python
# convert the column to numeric 
median_income['Medianeinkommen in EUR'] = pd.to_numeric(median_income['Medianeinkommen in EUR'], errors='coerce')
```

```python
# Join the median income data (RAUMID col) with the lor shapes (PLR_ID col)

lor = lor.merge(median_income, left_on="PLR_ID", right_on="RAUMID")
```

```python
lor.columns
```

```python
import matplotlib.pyplot as plt
# plot 'Medianeinkommen in EUR' column as choropleth map

plt.figure(figsize=FIGSIZE, dpi=DPI)
lor.plot(column='Medianeinkommen in EUR', cmap='viridis', legend=True, edgecolor='black', linewidth=0.2)
plt.title("Median Income in Berlin by LOR (2023)", fontsize=12)
plt.axis('off')
plt.savefig(REPORT_ROOT / "figures" / "median_income_map.png", bbox_inches='tight', transparent=True, dpi=DPI)
```
